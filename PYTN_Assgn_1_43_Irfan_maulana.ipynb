## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**
tampilkan data kejahatan minor ataupun mayor dengan diagram pie chart berdasarkan nama kejahatan dari tahun 2008 - 2016
(Make your own questions)

```python
# Write your function below
def pies (param1): # berikut fungsi yang saya gunakan untuk membuat gafik pie
    if param1 == 'minor_category' or param1 == 'major_category' : # ini untuk mengecek kategori
        dfs =  criminal.groupby(['year',param1]).size().reset_index(name="jml") # melakukan grouping serta menambahkan kolom jumlah dari hasil perhitungan berapa banayaknya data 
        dfs_pivot = dfs.pivot(index = param1, columns = 'year', values = 'jml') # melakukan perubahan data dengan fungsi pivot
        dfs_pivot['total'] = dfs_pivot.sum(axis=1) # Penambahan kolom dari penjumlahan data yang berbentuk int/float
        dfs_pivot['total'].plot(kind='pie',
                                      figsize=(15, 10),
                                      autopct='%1.1f%%',
                                      startangle=90,
                                      subplots=True,
                                      shadow=True,
                                      labels=None,         # turn off labels on pie chart
                                      # the ratio between the center of each pie slice and the start of the text generated by autopct
                                      pctdistance=1.12,
            #                           explode=explode_list  # 'explode'
                                      )
                # Graded-Funtion End

        # scale the title up by 12% to match pctdistance
        kategori = param1.replace('_',' ').title()
        plt.title('Data kriminal dengan kategori '+kategori+' berdasarkan nama kejahatan dari tahun 2008-2016 ', y=1.12)

        plt.axis('equal')

        # add legend
        plt.legend(labels=dfs_pivot.index, loc='upper left')

        plt.show()
    else:
        print("Harap input dangan salah satu kategori minor / mayor !!")
```

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**
Tampilkan data box plot setiap tahun
(Make your own questions)

```python
# Write your function below
def boxplot (param1): # berikut fungsi yang saya gunakan untuk membuat gafik box plot
    if param1 == 'minor_category' or param1 == 'major_category' : # ini untuk mengecek kategori
        df_minor = criminal.groupby(['year',param1]).size().reset_index(name="count");
        criminal_minor = df_minor.pivot(index = param1, columns = 'year', values = 'count').fillna(0);
        clmn=criminal_minor.columns.tolist()
        # Graded-Funtion Begin (~1 Lines)
        criminal_minor.boxplot(column=clmn,vert = False,figsize = (16, 8))
        # Graded-Funtion End
        kategori = param1.replace('_',' ').title()
        plt.title('repersentasi distribusi terkait data kejahatan '+kategori+' setiap tahun')
        plt.ylabel('Tahun')

        plt.show()
    else:
         print("Harap input dangan salah satu kategori minor / mayor !!")
```

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

bandingkan antaran kategori minor dan mayor di dari thn 2008 s/d 2016

```python
criminal_minor_category = criminal[['year','minor_category']]
criminal_major_category = criminal[['year','major_category']]


criminal_minor_category = criminal_minor_category.groupby(['year','minor_category']).size().reset_index(name="minor_category_count")
criminal_major_category = criminal_major_category.groupby(['year','major_category']).size().reset_index(name="major_category_count")

criminal_minor_category_pivot=criminal_minor_category.pivot(index='year',columns='minor_category',values='minor_category_count')
criminal_major_category_pivot=criminal_major_category.pivot(index='year',columns='major_category',values='major_category_count')

criminal_minor_category_pivot['total_minor']=criminal_minor_category_pivot.sum(axis=1)
criminal_major_category_pivot['total_major']=criminal_major_category_pivot.sum(axis=1)
criminal_minor_category_pivot = criminal_minor_category_pivot.drop(criminal_minor_category_pivot.columns.values[:-1], axis = 1)
criminal_major_category_pivot = criminal_major_category_pivot.drop(criminal_major_category_pivot.columns.values[:-1], axis = 1)
marges=pd.merge(criminal_major_category_pivot,criminal_minor_category_pivot, on="year", how = "right")
marges.reset_index(inplace=True)
criminal_major_category_pivot
# marges

norm_major = (marges['total_major'] - marges['total_major'].min()) / (marges['total_major'].max() - marges['total_major'].min())
norm_minor = (marges['total_minor'] - marges['total_minor'].min()) / (marges['total_minor'].max() - marges['total_minor'].min())
ax0=marges.plot(
    kind='scatter', 
    x='year', 
    y='total_major', 
    figsize=(10, 6), 
    alpha=0.5, 
    color='green',
    s = norm_major * 2000+10,
    xlim=(2008,2016)
)

ax1=marges.plot(
    kind='scatter', 
    x='year', 
    y='total_minor', 
    alpha=0.5, 
    color='blue',
    s = norm_minor * 2000+10,
    ax=ax0
)

ax0.set_ylabel('Angka kriminal')
ax0.set_title('Angka Kriminal major dan minor dari 2008 s/d 2016')
ax0.legend(['Major','Minor'],loc="upper left",fontsize='x-large')
```

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
criminal_minor_category = criminal[['year','minor_category']]
criminal_major_category = criminal[['year','major_category']]


criminal_minor_category = criminal_minor_category.groupby(['year','minor_category']).size().reset_index(name="minor_category_count")
criminal_major_category = criminal_major_category.groupby(['year','major_category']).size().reset_index(name="major_category_count")

criminal_minor_category_pivot=criminal_minor_category.pivot(index='year',columns='minor_category',values='minor_category_count')
criminal_major_category_pivot=criminal_major_category.pivot(index='year',columns='major_category',values='major_category_count')
marges=pd.merge(criminal_major_category_pivot,criminal_minor_category_pivot, on="year", how = "right")
marges_all = marges.transpose()
marges_all['total']=marges_all.sum(axis=1)
total_criminal = marges_all['total'].sum()
max_word=90
word_string =''
for cr in marges_all.index.values:
    if len(cr.split(' '))==1 :
        repeat_num_times = int(marges_all.loc[cr,'total']/float(total_criminal)*max_word)
        word_string = word_string +((cr + ' ')*repeat_num_times)
        
word_string
```

```python
# instantiate a word cloud object
your_wordcloud = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
your_wordcloud.generate(word_string)
```

```python
# Write your function below

# Graded-Funtion Begin (~1 Lines)
plt.imshow(your_wordcloud, interpolation='bilinear')
# Graded-Funtion End

plt.axis('off')
plt.show()
```
