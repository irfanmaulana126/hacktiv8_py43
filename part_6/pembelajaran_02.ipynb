```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'data/obes.xls'
data = pd.ExcelFile(path)
```

```python
data.sheet_names
```

```python
data_age = data.parse('7.2', skiprows=4,skipfooter=14)
```

```python
data_age.head()
```

```python
data_age.dropna(inplace=True)
```

```python
data_age.head()
```

```python
data_age.rename(columns={'Unnamed: 0':'Year'},inplace=True)
```

```python
data_age.tail()
```

```python
data_age.info()
```

```python
data_age['Year'].value_counts()
```

```python
data_age.set_index('Year',inplace=True)
```

```python
data_age.head()
```

```python
data_age.plot();
```

```python
data_age_minus_total = data_age.drop('Total',axis=1)
```

```python
data_age_minus_total.plot();
plt.title("Obes Analysis By Age");
plt.ylabel("Frequency");
```

```python
data_age_minus_total.plot(kind='hist');
plt.title("Obes Analysis By Age");
plt.ylabel("Frequency");
```

```python
selected_cols = ['Under 16','75 and over']
data_age_minus_total[selected_cols].plot();
plt.title("Obes Analysis By Age");
plt.ylabel("Frequency");
```
